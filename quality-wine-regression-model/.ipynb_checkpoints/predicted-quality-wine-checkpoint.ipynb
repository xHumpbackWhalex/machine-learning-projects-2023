Part 4: Linear Regression (4 Points)

```python
# Regression Models
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

#df.head()
```

```python
# list to store the results
method = []
r2_list = []
rmse_list = []
```

## Getting features and response variables

```python
response = df['quality']
features = df.drop('quality', axis=1)
```

## Split the data in trainig and test suset

```python
#Split the dataset in training and test
X_train, X_test, y_train, y_test = train_test_split(features, 
                                                    response, 
                                                    test_size=0.1, random_state=42)
```

## Training Linear Regression Model

```python
# create a linear regression model
model = LinearRegression()

#fit the model
model.fit(X_train, y_train)

#prediction test subset
pred = model.predict(X_test)

## Calulate the performance metrics
r2 = model.score(X_test, y_test)
print('R2: %.4f' % r2)

# Calculate RMSE
rmse = mean_squared_error(y_test, pred)**0.5
print('RMSE: %.4f' % rmse)

#Append the results
method.append('Linear Regression')
r2_list.append(r2)
rmse_list.append(rmse)
```

```python
from sklearn.linear_model import Ridge
y = df['quality']
X = df.drop('quality', axis=1)
#Split the dataset in training and test
X_train_rdg, X_test_rdg, y_train_rdg, y_test_rdg = train_test_split(X, y, test_size=0.1, random_state=42)

#alphas
alphas_rgd = [100, 10, 1, 0.1, 1e-2, 1e-4, 1e-6, 1e-8]

ridge = Ridge()
coefs = []

for a in alphas_rgd:
    ridge.set_params(alpha = a)
    ridge.fit(X,y)
    coefs.append(ridge.coef_)
np.shape(coefs)
```

```python
rmse_list_alpha = []
r2_list_alpha = []
r2_train_list_alpha = []

for a in alphas_rgd:
    ridge = Ridge(alpha = a, max_iter = 1000)
    ridge.fit(X_train_rdg, y_train_rdg)
    pred = ridge.predict(X_test_rdg)
    
    r2_train_list_alpha.append(ridge.score(X_train_rdg, y_train_rdg))
    r2_list_alpha.append(ridge.score(X_test_rdg,y_test_rdg))
    rmse_r = mean_squared_error(y_test_rdg, pred)**0.5
    rmse_list_alpha.append(rmse_r)
    
ridge_result = np.vstack((alphas_rgd, r2_train_list_alpha, r2_list_alpha, rmse_list_alpha)).T
ridge_df = pd.DataFrame(ridge_result, columns=['Alphas', 'R2 Train', 'R2 Test', 'RMSE'])
ridge_df.sort_values(by=['R2 Test'], ascending=False).head(3)
```

```python
## Find out the coefficient for the best result

ridge = Ridge(alpha = 1.000000e-02)
ridge_result = ridge.fit(X_train_rdg, y_train_rdg)
pred = ridge_result.predict(X_test_rdg)

r2_ridge = ridge_result.score(X_test_rdg, y_test_rdg)
rmse_ridge = mean_squared_error(y_test_rdg, pred)**0.5

#Append the results
method.append('Ridge')
r2_list.append(r2_ridge)
rmse_list.append(rmse_ridge)
```

```python
plt.figure(figsize=(7,5))
actual_values = y_test_rdg
plt.scatter(pred, actual_values, c='crimson')

p1 = max(max(pred), max(actual_values))
p2 = min(min(pred), min(actual_values))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('Actual values wine quality', fontsize=15)
plt.ylabel('Predictions values wine quality', fontsize=15)
plt.title('Predicted values vs. actual values wine quality')
plt.axis('equal')
plt.show()
```

## Lasso regression

```python
# Features and response variables
y = df['quality']
X = df.drop('quality', axis=1)

#Split the data in training and test subsets
X_train_lss, X_test_lss, Y_train_lss, Y_test_lss = train_test_split(X, y, 
                                                                    test_size = 0.1, 
                                                                    random_state = 42)

alphas_lss = np.linspace(1, 0.01, 20)

# Tuning alpha parameter in LASSO linear model
lasso = Lasso()
coefs = []

for a in alphas_lss:
    lasso.set_params(alpha = a)
    lasso.fit(X,y)
    coefs.append(lasso.coef_)

np.shape(coefs)
```

```python
rmse_list_ls = []
r2_list_ls = []
r2_train_list_ls = []

for a in alphas_lss:
    lasso = Lasso(alpha = a, max_iter=1000)
    lasso.fit(X_train_lss, Y_train_lss)
    pred = lasso.predict(X_test_lss)
    
    r2_train_list_ls.append(lasso.score(X_train_lss, Y_train_lss))
    r2_list_ls.append(lasso.score(X_test_lss,Y_test_lss))
    rmse_ls = mean_squared_error(Y_test_lss, pred)**0.5
    rmse_list_ls.append(rmse_ls)
    #print('RMSE: %.2f'% rmse)
    
lasso_result = np.vstack((alphas_lss, r2_train_list_ls, r2_list_ls, rmse_list_ls)).T
lasso_df = pd.DataFrame(lasso_result, columns=['Alphas','R2 Train', 'R2 Test', 'rmse'])
lasso_df.sort_values(by=['R2 Test'], ascending=False).head(3)
```

```python
# find out the coefficient of the best result obtained from our lasso

lasso = Lasso(alpha = 0.001)
lasso.fit(X_train_lss, Y_train_lss)
pred_lss = lasso.predict(X_test_lss)

r2_lasso = lasso.score(X_test_lss, Y_test_lss)
rmse_lasso =  mean_squared_error(Y_test_lss, pred_lss)**0.5

print('R2 in Test Set: %.4f'% lasso.score(X_test_lss, Y_test_lss))
print('RMSE in Test Set: %.4f' % mean_squared_error(Y_test_lss, pred_lss)**0.5)

# Append the results of Linear Regression Model
method.append('LASSO')
r2_list.append(r2_lasso)
rmse_list.append(rmse_lasso)
```

# Polynomial Regression Model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt
import numpy as np

# Features and response variables
y = df['quality']
X = df.drop('quality', axis=1)

# split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# create polynomial features
poly_reg = PolynomialFeatures(degree=2)
X_train_poly = poly_reg.fit_transform(X_train)
X_test_poly = poly_reg.transform(X_test)

# train a linear regression model on the polynomial features
model = LinearRegression()
model.fit(X_train_poly, y_train)

# predict on the test set and calculate scores
y_pred = model.predict(X_test_poly)
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
coefficient = model.coef_
intercept = model.intercept_
print("R^2 score:", r2)
print("RMSE score:", rmse)

# plot actual vs. predicted values
plt.scatter(y_test, y_pred)
plt.plot()
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.show()

# Append the results of Linear Regression Model
method.append('Polynomial')
r2_list.append(r2)
rmse_list.append(rmse)
```

- 2. Overfitting
when the model cannot generalize and fits too closely to the training dataset instead. Ridge regression adds an extra term to the cost function of linear regression. The purpose of adding these penalty terms is to ensure regularization, which shrinks the weights of the model to zero or close to zero. This helps prevent the model from overfitting the data.

__Overfitting__ occurs when a statistical model is trained too well on a specific dataset, to the point that it begins to capture noise or random fluctuations in the data instead of the underlying pattern. This results in a model that performs very well on the training data, but poorly on new or unseen data.

__Linear regression with Ridge__, also known as Ridge regression, is a regularization technique that can help solve the problem of overfitting. Ridge regression adds a penalty term to the least-squares objective function, which shrinks the coefficients towards zero and reduces the complexity of the model. This penalty term is controlled by a hyperparameter called alpha.

By adding the penalty term, Ridge regression effectively reduces the variance of the model, preventing it from overfitting the training data. It achieves this by sacrificing some bias in the model, which means that it may not fit the training data as well as a non-regularized linear regression model. However, the reduced variance generally leads to better performance on new or unseen data, making Ridge regression a useful technique for improving the generalization of linear regression models.


__Linear regression with Lasso__, also known as Lasso regression, is another regularization technique that can help solve the problem of overfitting in linear regression models.

Like Ridge regression, Lasso regression adds a penalty term to the least-squares objective function. However, instead of using the L2 norm penalty as in Ridge regression, Lasso uses the L1 norm penalty. This has the effect of shrinking some coefficients to zero, effectively performing feature selection and reducing the complexity of the model.

By reducing the number of features used in the model, Lasso regression can help prevent overfitting and improve the generalization of the model. Additionally, the sparsity introduced by Lasso can provide a more interpretable model by highlighting the most important features.

One advantage of Lasso regression over Ridge regression is that it can be particularly effective when there are a large number of features, some of which may be irrelevant or redundant. Lasso's feature selection capabilities can help identify and exclude these features, resulting in a more parsimonious and accurate model

RMSE is the average distance between the predicted values from the model and the actual values in the dataset. A lower RMSE indicates a better fitting model as it reflects a smaller error in our predictions.
r2 = the proportion of the variance in the response variable that can be explained by the predictor variable(s) in the model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt
import numpy as np

df2 = pd.DataFrame()
df2['quality'], df2['alcohol'] = df['quality'], df['alcohol']


X_train, X_test, y_train, y_test = train_test_split(df2['quality'], df2['alcohol'], random_state=42)

X_train_df, X_test_df = pd.DataFrame(X_train), pd.DataFrame(X_test)

poly = PolynomialFeatures(degree=2)
X_train_poly, X_test_poly = poly.fit_transform(X_train_df), poly.fit_transform(X_test_df)
#X_train_poly


model = LinearRegression()
model = model.fit(X_train_poly, y_train)

coefficient = model.coef_
intercept = model.intercept_
```

RMSE (Root Mean Squared Error) and R^2 (Coefficient of Determination) are two commonly used evaluation metrics in machine learning for assessing the performance of a linear regression model.

RMSE measures the average difference between the predicted values and the actual values in the dataset, taking into account the square of the differences between predicted and actual values. It gives an idea of how well the model's predictions match the actual data. A lower RMSE indicates a better fit of the model to the data.

R^2, on the other hand, measures the proportion of variation in the dependent variable that is explained by the independent variables in the model. It ranges between 0 and 1, where 0 indicates that the model explains none of the variation in the dependent variable, and 1 indicates that the model explains all the variation in the dependent variable. A higher R^2 indicates a better fit of the model to the data.

Both RMSE and R^2 are important in a linear regression model because they help to evaluate the accuracy and reliability of the model. A good model should have a low RMSE, indicating that the model's predictions are close to the actual values. At the same time, it should have a high R^2, indicating that a large proportion of the variation in the dependent variable is explained by the independent variables.

By using these two metrics together, a data scientist can get a better understanding of how well a linear regression model is performing and can make adjustments as necessary to improve its accuracy and reliability.
