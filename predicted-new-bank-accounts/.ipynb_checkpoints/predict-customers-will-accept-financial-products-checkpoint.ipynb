# 2. Library import and data loading

````python
import pandas as pd
import numpy as np
np.random.seed(42) 
import warnings
warnings.filterwarnings("ignore")

df = pd.read_csv('bank-sampled-9K.csv')
df = df.drop(['Unnamed: 0'], axis=1)
df.head(10)
````

# 3. Data Analysis, Preparing and Wrangling

## Dropping 'duration' column

````python
#drop duration as suggested by the documentation
df.drop('duration', axis=1, inplace=True)
````

## Convert the two-column time to a column containing

````python
# the number of days till the last day of campaign
from datetime import datetime as dt

last_date = "18 Nov, 2010" #assume the analysis date
df_1 = dt.strptime(last_date, "%d %b, %Y")

# the month needs to start with a capital letter
df['month'] = df['month'].apply(lambda x: x.capitalize())
df['date'] = df['day'].astype(str)+ " " + df['month'] + ", 2010"

# calculate the duration between the two dates
df['last_contact'] = df['date'].apply(lambda x: (df_1 - dt.strptime(x, "%d %b, %Y")).days)

df.drop(columns = {'day', 'date', 'month'}, axis=1, inplace=True)
````

## peek of the data

````python
df.head(5)
````

## Summary statistics of the data

````python
df.describe()
````

## Creating dummy variables

````python
df = pd.get_dummies(df, drop_first=True)
df.head()
````

````python
df.columns
````

````python
df.columns = df.columns.map(lambda x: x.replace('-', '_'))
df.columns
````

# 4. Feature Selecting and Scaling

## Define the features and target

````python
#Output variable (desired target): 
#y - has the client subscribed a term deposit? (binary: 'yes','no')
#after get the dummy variables the new variables is y_yes

from sklearn.model_selection import train_test_split

X = df.drop(['y_yes'], axis=1)#feature
y = df['y_yes']               #target
print('features: ', X.shape)
print('target:   ', y.shape)
````

## 4.1 Feature Selection Methods

````python
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler
from sklearn.svm import LinearSVC
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_classif
````

````python
feature_selec = []
shape_df = []

#Logitic Regression
logreg = SelectFromModel(estimator=LogisticRegression(max_iter=5000)).fit(X, y)
X_lr = X.iloc[:, logreg.get_support(indices=True)]
feature_selec.append("LogisticRegression")
shape_df.append(X_lr.shape[1])

#LinearSVC
lsvc = SelectFromModel(estimator=LinearSVC(C=0.001)).fit(X, y)
X_svc = X.iloc[:, lsvc.get_support(indices=True)]
feature_selec.append("LinearSVC")
shape_df.append(X_svc.shape[1])

#Featuring selection model with SelectKBest()
kbest = SelectKBest(score_func = mutual_info_classif, k=10).fit(X, y)
X_Kbest = X.iloc[:,kbest.get_support(indices=True)]
feature_selec.append("SelectKBest")
shape_df.append(X_Kbest.shape[1])

#Feature Selection results
result = np.vstack((feature_selec, shape_df)).T
result_df = pd.DataFrame(result, columns=['feature selection method', '# features'])
#result_df.sort_values(by=['# features'], ascending=False)
feature_selection_step_4 = result_df.sort_values(by=['# features'], ascending=False)
feature_selection_step_4
````

#### RESULT: The best of feature selection method was using Logistic Regression as the estimator with 7 features

## 5.2 Scaling the feature selected with RobustScaler 

````python
from sklearn.preprocessing import RobustScaler

X = feature1    #feature
y = df['y_yes'] #target

# Use robust scaler to scale the feature selected.
rs = RobustScaler()
zscore = rs.fit_transform(X)
zscore

#Feature scaled
feature_rs_lr = pd.DataFrame(zscore, index=X.index, columns = X.columns)
````

## 5.3 Classifier Modelling Pipeline

Use pipeline to implement the following classifier models:

-     Naïve Bayes classifier 
-     KNN
-     Decision tree classifier

````python
#first install xgboost
#!pip install xgboost 
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold

#Classifiers
names = ["Naïve Bayes clf", 
         "KNN clf", 
         "Decision tree clf", 
         "Random forest clf", 
         "Ada BOOST clf",
         "XGBOOST clf"]

scores = []

classifiers = [
    GaussianNB(),
    KNeighborsClassifier(),
    DecisionTreeClassifier(max_depth=5),
    RandomForestClassifier(n_estimators=100, max_depth=5),
    AdaBoostClassifier(),
    XGBClassifier()
]

for classifier in classifiers:
    pipeline = Pipeline(steps=[
        ('classifier', classifier)
    ])
    pipeline.fit(X_train, y_train)
    print(classifier)
    score = pipeline.score(X_test, y_test)
    scores.append(score)
    print("model score: ", score)
    print('--------------------------------')
````

## 5.5 Classifier Modelling Pipeline with k-fold cross validation

````python
# You need to use the scaled features, obtained from the previous step (5.2 )

X = feature1    #feature
y = df['y_yes'] #target

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.25, random_state=42)

#
scaler = RobustScaler()


GNB_clf = GaussianNB()
KNN_clf = KNeighborsClassifier()
dt_clf  = DecisionTreeClassifier()
rf_clf  = RandomForestClassifier()
ada_clf = AdaBoostClassifier()
xgb_clf = XGBClassifier()

#
GNB_pipeline = Pipeline([('scaler', scaler), ('classifier', GNB_clf)])
KNN_pipeline = Pipeline([('scaler', scaler), ('classifier', KNN_clf)])
dt_pipeline = Pipeline([('scaler', scaler), ('classifier', dt_clf)])
rf_pipeline = Pipeline([('scaler', scaler), ('classifier', rf_clf)])
ada_pipeline = Pipeline([('scaler', scaler), ('classifier', ada_clf)])
xgb_pipeline = Pipeline([('scaler', scaler), ('classifier', xgb_clf)])

# Step 4: Define the parameters for each classifier
GNB_params = {}
KNN_params = {}
dt_params = {'classifier__max_depth': [5]}
rf_params = {'classifier__n_estimators': [100], 'classifier__max_depth': [5]}
adaboost_params = {}
xgb_params = {}


#Define the pipelines and their respective parameters
pipelines = {"Naïve Bayes clf": [GNB_pipeline, GNB_params],
             "KNN clf": [KNN_pipeline, KNN_params],
             "Decision tree clf": [dt_pipeline, dt_params],
             "Random forest clf": [rf_pipeline, rf_params],
             "Ada BOOST clf": [ada_pipeline, adaboost_params],
             "XGBOOST clf": [xgb_pipeline, xgb_params]
            }

# Step 6: Loop over each pipeline and their respective parameters, and perform k-fold cross-validation on the training set

kfold = KFold(n_splits=5, shuffle=True, random_state=42)
cv_results = {}
best_score_1 = 0
for pipeline_name, pipeline_obj in pipelines.items():
    pipeline, params = pipeline_obj[0], pipeline_obj[1]
    grid_search = GridSearchCV(pipeline, params, cv=kfold)
    grid_search.fit(X_train, y_train)
    cv_results[pipeline_name] = grid_search.best_score_ #Mean cross-validated score of the best_estimator
    if grid_search.best_score_>best_score_1:
        best_score_1 = grid_search.best_score_
        grid_search_1 = grid_search
````

````python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Lasso, Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
import numpy as np

# Create pipeline with StandardScaler and Lasso regression
lasso_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('lasso', Lasso())
])

# Set hyperparameters for Lasso regression
lasso_params = {
    'lasso__alpha': np.logspace(-3, 3, 7),
    'lasso__max_iter': [1000, 5000, 10000]
}

# Use GridSearchCV to find the best hyperparameters for Lasso regression
lasso_grid = GridSearchCV(lasso_pipe, lasso_params, cv=5, scoring='neg_mean_squared_error')
lasso_grid.fit(X_train, y_train)

# Create pipeline with StandardScaler and Ridge regression
ridge_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('ridge', Ridge())
])

# Set hyperparameters for Ridge regression
ridge_params = {
    'ridge__alpha': np.logspace(-3, 3, 7),
    'ridge__max_iter': [1000, 5000, 10000]
}

# Use GridSearchCV to find the best hyperparameters for Ridge regression
ridge_grid = GridSearchCV(ridge_pipe, ridge_params, cv=5, scoring='neg_mean_squared_error')
ridge_grid.fit(X_train, y_train)
````

# 7.3.2 How can we drop the value of false negatives
The high number of false negatives in the classification results could be due to class imbalance in the dataset. Addressing class imbalance by undersampling or oversampling, or using algorithms specifically designed for imbalanced datasets like SMOTE, can help to balance the dataset and prevent classifier bias.

To improve the classifier performance, hyperparameter tuning can also be used to optimize the model for the specific dataset. Feature engineering can also extract more relevant features from the data to improve the performance of the classifier. By addressing class imbalance and optimizing the model and features, the classifier can improve its ability to predict the minority class and reduce the number of false negatives.

# 5. To try different models with different scalers

````python
#RobustScaler, StandardScaler and Minmax scalre result in same score
    
X = df.drop(["y_yes"],axis=1) #feature
y = df['y_yes'] #target

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.25, random_state=42)



scalers = {"RobustScaler": RobustScaler(),
            "StandardScaler": StandardScaler(),
            "MinMaxScaler": MinMaxScaler(),
            "MaxAbsScaler": MaxAbsScaler(),
            "PowerTransformer": PowerTransformer(),
            "QuantileTransformer": QuantileTransformer()           
                    }


best_score_1 = 0
for scaler_name, scaler in scalers.items():
    GNB_clf = GaussianNB()
    KNN_clf = KNeighborsClassifier()
    dt_clf  = DecisionTreeClassifier()
    rf_clf  = RandomForestClassifier()
    ada_clf = AdaBoostClassifier()
    xgb_clf = XGBClassifier()
    log_clf = LogisticRegression()

    
    
    #
    GNB_pipeline = Pipeline([('scaler', scaler), ('classifier', GNB_clf)])
    KNN_pipeline = Pipeline([('scaler', scaler), ('classifier', KNN_clf)])
    dt_pipeline = Pipeline([('scaler', scaler), ('classifier', dt_clf)])
    rf_pipeline = Pipeline([('scaler', scaler), ('classifier', rf_clf)])
    ada_pipeline = Pipeline([('scaler', scaler), ('classifier', ada_clf)])
    xgb_pipeline = Pipeline([('scaler', scaler), ('classifier', xgb_clf)])
    log_pipeline = Pipeline([('scaler', scaler), ('classifier', log_clf)])


    GNB_params = {}
    KNN_params = {}
    dt_params = {'classifier__max_depth': [5]}
    rf_params = {'classifier__n_estimators': [100], 'classifier__max_depth': [5]}
    adaboost_params = {}
    xgb_params = {}
    log_params = {}


    pipelines = {"Naïve Bayes clf": [GNB_pipeline, GNB_params],
                 "KNN clf": [KNN_pipeline, KNN_params],
                 "Decision tree clf": [dt_pipeline, dt_params],
                 "Random forest clf": [rf_pipeline, rf_params],
                 "Ada BOOST clf": [ada_pipeline, adaboost_params],
                 "XGBOOST clf": [xgb_pipeline, xgb_params],
                 "Logistic clf": [log_pipeline, log_params]
                }

    # Loop over each pipeline and their respective parameters, and perform k-fold cross-validation on the training set

    kfold = KFold(n_splits=10, shuffle=True, random_state=42)
    cv_results = {}

    for pipeline_name, pipeline_obj in pipelines.items():
        pipeline, params = pipeline_obj[0], pipeline_obj[1]
        grid_search = GridSearchCV(pipeline, params, cv=kfold)
        grid_search.fit(X_train, y_train)
        cv_results[pipeline_name] = grid_search.best_score_ #Mean cross-validated score of the best_estimator
        if grid_search.best_score_> best_score_1:
            best_score_1 = grid_search.best_score_
            grid_search_1 = grid_search

    sorted_cv_results_step_5 = dict(sorted(cv_results.items(), key=lambda item: item[1], reverse=True))
    sorted_cv_results_step_5
    print("\n Best Scaler: ", scaler_name)
    print()
    print(sorted_cv_results_step_5)
````
